```python
def fits_to_image(file_name):
    if not os.path.isfile(r"C:\Users\deych\Desktop\Desktop Old\Summer Projects\SRFP\Python and Jupyter\Images\\"+file_name[:-5]+'_image.jpeg'):
        with fits.open(path+'\\'+file_name) as f:
            primary_data = f[0].data
        data = primary_data.copy()
        bias = 1
        for i in range(len(data)):
            for j in range(len(data[i])):
                if data[i][j]<bias:
                    data[i][j] = bias
        plt.figure(figsize = (8,8))
        plt.imshow(data,cmap = 'hot',norm = colors.LogNorm(vmin=data.min(), vmax=data.max()))
        plt.title('Frame = '+str(file_name[:-5]))
        plt.savefig(r"C:\Users\deych\Desktop\Desktop Old\Summer Projects\SRFP\Python and Jupyter\Images\\"+file_name[:-5]+'_image.jpeg')
```
