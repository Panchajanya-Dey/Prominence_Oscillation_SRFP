```python
import os
target_path = r"C:\Users\deych\Desktop\Desktop Old\Summer Projects\SRFP\Python and Jupyter\FITS Files\data14"
```

```python
import shutil

def copy_file(source_path, destination_path):
    shutil.copy(source_path, destination_path)
    print(f"File copied from {source_path} to {destination_path}")

def delete_file(file_path):
    if os.path.exists(file_path):
        os.remove(file_path)
        print(f"File '{file_path}' deleted successfully.")
    else:
        print(f"File '{file_path}' does not exist.")
```

```python
files = []
contents = os.listdir(target_path)
n = 0
for file_name in contents:
    if os.path.isfile(os.path.join(target_path, file_name)):
        print(str(n)+" : "+file_name)
        n+=1
        files.append(file_name)
```

```python
n
```

```python
num_file = int(n/16)
num_file
```

```python
final_path = r"C:\Users\deych\Desktop\Desktop Old\Summer Projects\SRFP\Python and Jupyter\FITS Files\data14\data"
for i in range(15):
    contents = os.listdir(target_path)
    path = final_path+str(i)
    for j in range(num_file):
        file = target_path+'\\'+contents[j]
        copy_file(file,path)
        delete_file(file)
```
